#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

population_filter = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

filtered_data = ds[ds['level5'].isin(population_filter)].copy()

# Normalizaciñón del precio para las diferentes poblaciones:

filtered_data['normalized_price'] = filtered_data.groupby('level5')['price'].transform (lambda x: (x - x.min()) / (x.max() - x.min()))

# Gráfica
plt.figure(figsize=(12, 8))

# Histogramas para cada población:
for city in population_filter:
    plt.hist(filtered_data[filtered_data['level5'] == city]['normalized_price'], bins=10, alpha=0.5, label=city)

# Labesls
plt.title('Precio Normalizado por población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(title='Poblaciones')
plt.grid(axis='y')

plt.show()
```

```python
## HERE: plot the coordinates of the estates

import pandas as pd
from ipyleaflet import Map, CircleMarker, basemaps
from IPython.display import display

# Data sample:
data = {
    'level5': ['Getafe', 'Alcorcón', 'Móstoles', 'Leganés'],
    'latitude': [40.3083, 40.3453, 40.3223, 40.3272],
    'longitude': [-3.7327, -3.8278, -3.8645, -3.7635]
}
cinturon_sur_madrid = pd.DataFrame(data)

# Coodinates of Madrid:
madrid_center = (40.4168, -3.7038)
map = Map(center=madrid_center, zoom=10, basemap=basemaps.OpenStreetMap.Mapnik)

# Colors for each population:
colores = {
    'Getafe': 'red',
    'Alcorcón': 'blue',
    'Móstoles': 'green',
    'Leganés': 'orange'
}

# Adding marker locations in the map:
for index, row in cinturon_sur_madrid.iterrows():
    population = row['level5']
    lat = row['latitude']
    lon = row['longitude']
    if population in colores: 
        color = colores[population]
        marker = CircleMarker(location=(lat, lon),
                              radius=10,
                              color=color,
                              fill_color=color,
                              fill_opacity=0.6)
        map.add_layer(marker)

# Display the map
display(map)
```
